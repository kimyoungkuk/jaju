```python
featurelist = []
```

```python
start = time.time()

x=0
while x < len(faces.index):
    
    if(x%1000==0):
        print(x)
        
    featurebuf=list(faces.loc[x,faces.loc[x,:]==1].index)
    if(list(faces.loc[x,faces.loc[x,:]==1].index).count('the male')!=0):
        featurebuf.remove('the male')
    else:
        featurebuf.remove('the female')
        
    if(len(featurebuf)==0):
#         print(faces.loc[x,faces.loc[x,:]==1])
#         print(list(faces.loc[x,faces.loc[x,:]==1].index))
        featurelist.append(featurebuf)
#         print(featurelist)
#         print(featurelist[-1])
#         print(len(list(faces.loc[x,faces.loc[x,:]==1].index)))
#         print(rd)
    elif (len(featurebuf)<=1):
        rd=1
#         print(faces.loc[x,faces.loc[x,:]==1])
#         print(list(faces.loc[x,faces.loc[x,:]==1].index))
        featurelist.append(random.sample(featurebuf, rd))
#         print(featurelist)
#         print(featurelist[-1])
#         print(len(list(faces.loc[x,faces.loc[x,:]==1].index)))
#         print(rd)
    else:
        rd=random.randrange(1,len(featurebuf)+1)
        if(4<rd):
            rd=4
#         print(faces.loc[x,faces.loc[x,:]==1])
#         print(list(faces.loc[x,faces.loc[x,:]==1].index))
        featurelist.append(random.sample(featurebuf, rd))
#         print(featurelist)
#         print(featurelist[-1])
#         print(len(list(faces.loc[x,faces.loc[x,:]==1].index)))
#         print(rd)
        
    
#     print(len(list(faces.loc[x,faces.loc[x,:]==1].index)))

    if(list(faces.loc[x,faces.loc[x,:]==1].index).count('the male')!=0):
        featurelist[x].append('the male')
    else:
        featurelist[x].append('the female')
        
#     print(featurelist)
#     print(faces.loc[x])
    x=x+1
    
end = time.time()
print(end-start)
```

```python
sentencelist=[]

x=0
while x < len(faces.index):
# while x < 10:
    
    num_has=0
    for ifeat in featurelist[x][0:len(featurelist[x])-1]:
#         print(ifeat)
        if(ifeat[0]=='h'):
            num_has=num_has+1
            
#     print(len(featurelist[x])-1)
#     print(num_has)
    
    sentence=[]
    if(featurelist[x][-1]=='the male'):
        sentence.append('man')
    else:
        sentence.append('woman')
#     print(sentence)
    
    featloca=""
    if(len(featurelist[x])-1==4):
        if(num_has==4):
            featloca="2233"
        elif(num_has==3):
            featloca=random.choice(["2231","2331"])
            featloca="2331"
        elif(num_has==2):
            featloca="2231"
        elif(num_has==1):
            featloca="2331"
        else:
            featloca="3311"
    elif(len(featurelist[x])-1==3):
        if(num_has==3):
            featloca=random.choice(["223","233"])
        elif(num_has==2):
            featloca=random.choice(["223","231","331"])
        elif(num_has==1):
            featloca=random.choice(["233","231","311"])
        else:
            featloca=random.choice(["311","331"])
    elif(len(featurelist[x])-1==2):
        if(num_has==2):
            featloca=random.choice(["23","33"])
        elif(num_has==1):
            featloca=random.choice(["23","31"])
        else:
            featloca="31"
    elif(len(featurelist[x])-1==1):
        if(num_has==1):
            featloca="3"
        else:
            featloca="3"
    else:
        featloca=""
        
#     print(featloca)
    fn1=0
    fn2=0
    fn3=0
    loca1=0
    loca2=1
    loca3=1
    for i, ifeat in enumerate(featurelist[x][0:len(featurelist[x])-1],0):
        if(i<num_has):
            if(featloca[i]=='3'):
                if(fn3==0):
                    sentence.insert(loca3,ifeat)
                    fn3=fn3+1
                    loca3=loca3+1
                else:
                    sentence.insert(loca3,ifeat.replace('has','and'))
                    fn3=fn3+1
                    loca3=loca3+1
            else:
                if(fn2==0):
                    sentence.insert(loca2,ifeat.replace('has','with'))
                    fn2=fn2+1
                    loca2=loca2+1
                    loca3=loca3+1
                else:
                    sentence.insert(loca2,ifeat.replace('has','and'))
                    fn2=fn2+1
                    loca2=loca2+1
                    loca3=loca3+1
            
        else:
            if(featloca[i]=='3'):
                if(fn3==0):
                    sentence.insert(loca3,ifeat)
                    fn3=fn3+1
                    loca3=loca3+1
                else:
                    sentence.insert(loca3,ifeat.replace('is','and'))
                    fn3=fn3+1
                    loca3=loca3+1
            else:
                if(fn1==0):
                    sentence.insert(loca1,ifeat.replace('is ',''))
                    fn1=fn1+1
                    loca1=loca1+1
                    loca2=loca2+1
                    loca3=loca3+1
                else:
                    sentence.insert(loca1,ifeat.replace('is','and'))
                    fn1=fn1+1
                    loca1=loca1+1
                    loca2=loca2+1
                    loca3=loca3+1
                
    sentence.insert(0,'A')
#     print(sentence)
    sentencestr=" ".join(sentence)
    sentencelist.append(sentencestr)
    x=x+1
    
sentencelist
```
